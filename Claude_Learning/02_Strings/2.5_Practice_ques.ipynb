Q6)Build an interactive program that:
1]Asks the user for:patient name, age, medicine name, dose in mg, and how many times per day
2]Calculates the total daily dose (dose*times per day)
3]Prints a clean, formatted prescription.

```python
# ===== User Input =====
name = input("Enter patient name: ")
age = input("Enter patient age: ")
medicine = input("Enter medicine name: ")
dose = int(input("Enter dose in mg: "))
times = int(input("How many times per day:"))

# ===== Calculation =====
total_dose = dose*times

# ===== Clean the name =====
name = name.strip().title()
medicine = medicine.strip().title()

# ===== Print Precription =====
print()
print("="*40)
print("            PRESCRIPTION SLIP")
print("="*40)
print(f"   Patient Name : {name}")
print(f"   Age          : {age} years")
print(f"   Medicine     : {medicine}")
print(f"   Dose         : {dose}mg * {times} times/day")
print(f"   Total daily  : {total_dose}mg")
print(f"="*40)
print(f"    Please take medicine after meals")
print(f"="*40)
```
